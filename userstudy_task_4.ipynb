
# Task 4.1 Unsupported Methods
If you have played around with Plotserializer in the other task, you may have come across some of the Warnings.

Please take a look at the warnings below.
As of this moment, we do not support the method .boxplot().
You can still use the method, but the data will not be serialized!

> Do not try to fix Warning like **"is not supported by PlotSerializer!"** as they just tell you that the current methode is not yet supported by Plotserializer

> You can find the Plotserializer documentation [here](https://mt.gg/temp_plotserializer_docs/).

```python
from plot_serializer.matplotlib.serializer import MatplotlibSerializer
serializer = MatplotlibSerializer()
# ---

import matplotlib.pyplot as plt
import numpy as np

from matplotlib.patches import Polygon

# example taken from matplotlib.org
# Fixing random state for reproducibility
np.random.seed(19680801)

# fake up some data
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))

fig, ax = serializer.subplots()

ax.boxplot(data, 0, '')
ax.set_title("The Boxplot")

plt.show()

# ---
serializer.write_json_file("userstudy.json")
```
